```python
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
from src.read_write import load_json
```

```python
# Run this script from the thesis_coding/ dir
DATA_DIR = Path().resolve() / 'data/multi_task_learning/processed'
non_converged_runs = ['/4UHF_ICM1_ELCB3/exp_1.json']

for run in non_converged_runs:
    data = load_json(DATA_DIR, run)
```

```python
gmp = np.array(data['gmp'])[:,-2]
gmp_xy = np.array(data['gmp'])[:, 0:2]
xy = np.array(data['xy'])
```

```python
plt.plot(gmp)
plt.ylim(-0.5, 1.5)
plt.axhline(0.23)
plt.axhline(-0.23)
plt.show()
fig, axs = plt.subplots(1, 1, figsize=(13,5))
plt.plot(gmp_xy[:,0])
plt.plot(gmp_xy[:,1])
plt.scatter(np.arange(len(xy[:,0])), xy[:,0])
plt.scatter(np.arange(len(xy[:,0])), xy[:,1])
plt.show()
```
